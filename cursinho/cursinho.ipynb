```python
import pandas as pd

arquivo = pd.read_excel("cursinho.xlsx")
display(arquivo) 
```

```python
import matplotlib.pyplot as plt

setor = arquivo["Dias da semana"]
gasto = arquivo["Alunos com falta"]

plt.title("Número de alunos com faltas")
plt.pie(gasto, labels=setor, autopct="%1.2f%%")
plt.show()
```

Ao analisar o gráfico podemos ver que nos dias de sábado tem mais faltas que o resto dos dias, assim como sexta feira.
